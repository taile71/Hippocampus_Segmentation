## Loading NIFTI images using NiBabel

NiBabel is a python library for working with neuro-imaging formats (including NIFTI) that we have used in some of the exercises throughout the course. Our volumes and labels are in NIFTI format, so we will use nibabel to load and inspect them.

NiBabel documentation could be found here: https://nipy.org/nibabel/

Our dataset sits in two directories - *images* and *labels*. Each image is represented by a single file (we are fortunate to have our data converted to NIFTI) and has a corresponding label file which is named the same as the image file.

Note that our dataset is "dirty". There are a few images and labels that are not quite right. They should be quite obvious to notice, though. The dataset contains an equal amount of "correct" volumes and corresponding labels, and you don't need to alter values of any samples in order to get the clean dataset.

```python
img_path = "/data/TrainingSet/images"
lbl_path = "/data/TrainingSet/labels"
```

```python
# TASK: Your data sits in directory /data/TrainingSet.
# Load an image and a segmentation mask into variables called image and label

image = nib.load('/data/TrainingSet/images/hippocampus_001.nii.gz')
label = nib.load('/data/TrainingSet/labels/hippocampus_001.nii.gz')
                
```

```python
# Nibabel can present your image data as a Numpy array by calling the method get_fdata()
# The array will contain a multi-dimensional Numpy array with numerical values representing voxel intensities. 
# In our case, images and labels are 3-dimensional, so get_fdata will return a 3-dimensional array. You can verify this
# by accessing the .shape attribute. What are the dimensions of the input arrays?

# TASK: using matplotlib, visualize a few slices from the dataset, along with their labels. 
# You can adjust plot sizes like so if you find them too small:
# plt.rcParams["figure.figsize"] = (10,10)

image_slices = image.get_fdata()
lbl_slices = label.get_fdata()

print("Image shape: ", image_slices.shape)
print("Label shape: ", lbl_slices.shape)

```

```python
print("Viewing a few Axial image and label slices")

fig, m_axs = plt.subplots(8, 2, figsize = (5,15))
ax = m_axs.flatten()

for i in range(len(ax)):
    n_slice = (2*i)+3
    if i%2 ==0:
        ax[i].axis("off")
        ax[i].set_title("Axial-Image-slice: " + str(n_slice))
        ax[i].imshow(image_slices[:,:,n_slice],cmap='gray',aspect =1)
        ax[i+1].axis("off")
        ax[i+1].set_title("Axial-Label-slice: " + str(n_slice))
        ax[i+1].imshow(lbl_slices[:,:,n_slice],cmap='gray',aspect = 1)
        
plt.show()

            
```

```python
print("Viewing a few Sagittal image and label slices")

fig, m_axs = plt.subplots(8, 2, figsize = (5,15))
ax = m_axs.flatten()

for i in range(len(ax)):
    n_slice = (2*i)+4
    if i%2 ==0:
        ax[i].axis("off")
        ax[i].set_title("Sagittal-Image-slice: " + str(n_slice))
        ax[i].imshow(image_slices[n_slice,:,:],cmap='gray',aspect =1)
        ax[i+1].axis("off")
        ax[i+1].set_title("Sagittal-Label-slice: " + str(n_slice))
        ax[i+1].imshow(lbl_slices[n_slice,:,:],cmap='gray',aspect = 1)
        
plt.show()
```

```python
print("Viewing a few Corona image and label slices")

fig, m_axs = plt.subplots(12, 2, figsize = (5,20))
ax = m_axs.flatten()

for i in range(len(ax)):
    n_slice = (2*i)+4
    if i%2 ==0:
        ax[i].axis("off")
        ax[i].set_title("Corona-Image-slice: " + str(n_slice))
        ax[i].imshow(image_slices[:,n_slice,:],cmap='gray',aspect =1)
        ax[i+1].axis("off")
        ax[i+1].set_title("Corona-Label-slice: " + str(n_slice))
        ax[i+1].imshow(lbl_slices[:,n_slice,:],cmap='gray',aspect = 1)
        
plt.show()
```

## Looking at single image data
In this section we will look closer at the NIFTI representation of our volumes. In order to measure the physical volume of hippocampi, we need to understand the relationship between the sizes of our voxels and the physical world.

```python
# Nibabel supports many imaging formats, NIFTI being just one of them. I told you that our images 
# are in NIFTI, but you should confirm if this is indeed the format that we are dealing with
# TASK: using .header_class attribute - what is the format of our images?

print("Image header_class", image.header_class)
print("Label header_class", label.header_class)
```

```python
image_info = image.header
print(image_info)
```

Further down we will be inspecting .header attribute that provides access to NIFTI metadata. You can use this resource as a reference for various fields: https://brainder.org/2012/09/23/the-nifti-file-format/

```python
# TASK: How many bits per pixel are used?
print(image_info['bitpix'])
```

```python
# TASK: What are the units of measurement?
print(image_info['xyzt_units'])
```

UoM xyzt_units 10 (binary) is translated to 2 decimal which is milimeter(mm) per https://brainder.org/2012/09/23/the-nifti-file-format/

```python
# TASK: Do we have a regular grid? What are grid spacings?
print(image_info["pixdim"])
```

The grid is regular with spacing of 1.

```python
# TASK: What dimensions represent axial, sagittal, and coronal slices? How do you know?

print("Dimensions {} \n".format(image_info["dim"]))

print("X: {} From left towards [R]ight".format(image_info["srow_x"]))
print("Y: {} From  posterior to [A]nterior".format(image_info["srow_y"]))
print("Z: {} From inferior to [S]uperior".format(image_info["srow_z"]))
```

Based on the article https://theaisummer.com/medical-image-coordinates/, the 3D Slicer and srow Nibabel header info, the coordinate dimensions X,Y,and Z correspond to Sagittal, Coronal and Axial respectively.

```python
# By now you should have enough information to decide what are dimensions of a single voxel
# TASK: Compute the volume (in mm³) of a hippocampus using one of the labels you've loaded. 
# You should get a number between ~2200 and ~4500



#The volume of one voxel= pixdim[1] * pixdim[2] * pixdim[3] = 1.0 mm^3

print("Volume of a hippocampus label is {} mm^3".format(np.sum(lbl_slices > 0)))

```

### Some volume renderings

```python
#Orthograhpic Projections on Axial
figure, (ax0,ax1) = plt.subplots(1,2, figsize=(10,10))
img_reg = nib.load('/data/TrainingSet/images/hippocampus_242.nii.gz').get_fdata()

vr = np.zeros((img_reg.shape[0], img_reg.shape[1]))
for z in range (img_reg.shape[2]):
    vr += img_reg[:,:,z]

ax0.set_title("Axial Orthograhic Projections")
ax0.imshow(vr, cmap="gray")

#Maximum Intensity Projection
mip = np.zeros((img_reg.shape[0], img_reg.shape[1]))

for z in range(img_reg.shape[2]):
    mip = np.maximum(mip,img_reg[:,:,z])

ax1.set_title("Maximum Intensity Projection")
ax1.imshow(mip, cmap="gray")

plt.show()
```

### Examing outliers from Pix/Dimension and Unit of Measurements

```python
#Checking the training image files
img_dim_Info = []
img_files = sorted([p for p in pathlib.Path(img_path).iterdir() if p.is_file()])
for pf in img_files:
    img = nib.load(pf)
    image_info = img.header
    #lbl_dim_Info.append(image_info["dim"])
    if ((image_info["dim"][1] > 80) or (image_info["xyzt_units"] !=10) or (image_info["pixdim"][0] !=1)):
        print("Index: {} , file name: {} ".format(img_files.index(pf), pf))
        print("UoM ", image_info['xyzt_units'])
        print("pixdim ", image_info["pixdim"])
        print("dim",image_info["dim"])
        
```
